```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Step 1: Perform Mapping
occupation_mapping = {"Software Developer": 1, "Psychologist": 2, "Doctor": 3, "Teacher": 4, "Artist": 5, "Business Analyst": 6,
                      "Mechanical Engineer": 7, "Lawyer": 8, "Biologist": 9, "Economist": 10}
fields_mapping = {'Medicine': 1, 'Business': 2, 'Biology': 3, 'Education': 4, 'Mechanical Engineering': 5, 'Law': 6,
                   'Computer Science': 7, 'Arts': 8, 'Economics': 9, 'Psychology': 10}
gender_mapping = {"Male": 1, "Female": 2}
education_mapping = {"High School": 1, "Bachelor's": 2, "Master's": 3, "PhD": 4}
IGR_mapping = {"Low": 1, "Medium": 2, "High": 3}
family_mapping = {"Low": 1, "Medium": 2, "High": 3}

data['Current Occupation'] = data['Current Occupation'].map(occupation_mapping)
data['Field of Study'] = data['Field of Study'].map(fields_mapping)
data['Gender'] = data['Gender'].map(gender_mapping)
data['Education Level'] = data['Education Level'].map(education_mapping)
data['Industry Growth Rate'] = data['Industry Growth Rate'].map(IGR_mapping)
data['Family Influence'] = data['Family Influence'].map(family_mapping)

print("Mapped Data (First 5 rows):")
print(data.head())
```
