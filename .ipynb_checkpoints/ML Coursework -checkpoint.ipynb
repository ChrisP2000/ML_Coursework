```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load Data
# Assuming `data` is your DataFrame containing the World Happiness data.

# Step 2: Exclude Non-Numeric Columns
numeric_columns = ['GDP per capita', 'Social support', 'Healthy life expectancy', 
                   'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']
numeric_data = data[numeric_columns]

# Step 3: Drop Rows with Missing Values
numeric_data = numeric_data.dropna()
# Save the corresponding indices (for proper alignment)
indices = numeric_data.index

# Step 4: Normalize the Data
scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(numeric_data)

# Step 5: Apply K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
cluster_labels = kmeans.fit_predict(normalized_data)

# Step 6: Assign Clusters to the Original DataFrame
# Use the saved indices to align the cluster labels
data.loc[indices, 'Cluster'] = cluster_labels

# Step 7: Analyze Clusters
cluster_summary = data.groupby('Cluster')[numeric_columns].mean()
print("Cluster Summary:")
print(cluster_summary)

cluster_mapping = {
    0: "Least Developed Countries",
    1: "Developing Countries",
    2: "Developed Countries"
}

data['Cluster Label'] = data['Cluster'].map(cluster_mapping)

# Step 8: Visualize Clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x=normalized_data[:, 0], 
    y=normalized_data[:, 1], 
    hue=data.loc[indices, 'Cluster Label'],  # Use labeled clusters for visualization
    palette='viridis',
    s=100
)

plt.title('Clusters of Countries Based on World Happiness Data')
plt.xlabel('Normalized GDP per capita')
plt.ylabel('Normalized Social support')
plt.legend(title='Cluster')
plt.show()

# Step 9: Label Clusters
for cluster in cluster_summary.index:
    print(f"Cluster {cluster} Characteristics:")
    print(cluster_summary.loc[cluster])
    print()
    
    
print("Cluster Summary with Labels:")
for cluster, label in cluster_mapping.items():
    print(f"{label} (Cluster {cluster}):")
    print(cluster_summary.loc[cluster])
    print()

print("\nCluster Assignments (First 5 Rows):")
print(data[['Country or region', 'Cluster Label']].head())
```

```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Apply DBSCAN with adjusted parameters
dbscan = DBSCAN(eps=0.2, min_samples=5)  # Adjust eps and min_samples for stricter clustering
dbscan_labels = dbscan.fit_predict(normalized_data)

# Step 2: Define custom colors for clusters and outliers
custom_palette = {
    -1: "red",   # Outliers
    0: "blue",   # Cluster 0
}

# Step 3: Map cluster labels to the custom palette
mapped_labels = [custom_palette[label] if label in custom_palette else "gray" for label in dbscan_labels]

# Step 4: Visualize Clusters with GDP and Corruption
plt.figure(figsize=(10, 6))
for label in custom_palette:
    # Select points belonging to the current label
    cluster_points = normalized_data[dbscan_labels == label]
    plt.scatter(
        cluster_points[:, 0],  # GDP per capita
        cluster_points[:, 5],  # Perceptions of corruption
        c=custom_palette[label], 
        label=f"Cluster {label}" if label != -1 else "Outliers",
        s=100
    )

plt.title('DBSCAN Clustering on World Happiness Data (GDP and Corruption)')
plt.xlabel('Normalized GDP per capita')
plt.ylabel('Normalized Perceptions of Corruption')
plt.legend(title='DBSCAN Cluster')
plt.show()




```

```python
from sklearn.decomposition import PCA

# Step 1: Apply PCA to reduce to 2 components
pca = PCA(n_components=2)
pca_data = pca.fit_transform(normalized_data)

# Step 2: Visualize PCA-reduced data
plt.figure(figsize=(10, 6))
plt.scatter(pca_data[:, 0], pca_data[:, 1], c=dbscan_labels, cmap='viridis', s=100)
plt.title('DBSCAN Clustering after PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster')
plt.show()
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Step 1: Apply PCA to reduce to 2 components
pca = PCA(n_components=2)
pca_data = pca.fit_transform(normalized_data)

# Step 2: Define descriptive labels for clusters
cluster_mapping = {
    -1: "Outliers",
    0: "Cluster 0 - Developed Countries",
    1: "Cluster 1 - Developing Countries",
    2: "Cluster 2 - Least Developed Countries"
}

# Map DBSCAN labels to descriptive names
descriptive_labels = [cluster_mapping[label] if label in cluster_mapping else f"Cluster {label}" for label in dbscan_labels]

# Step 3: Visualize PCA-reduced data with descriptive cluster labels
plt.figure(figsize=(10, 6))
scatter = plt.scatter(
    pca_data[:, 0], 
    pca_data[:, 1], 
    c=dbscan_labels, 
    cmap='viridis', 
    s=100
)

# Add color bar with cluster descriptions
cbar = plt.colorbar(scatter, ticks=sorted(cluster_mapping.keys()))
cbar.ax.set_yticklabels([cluster_mapping[key] for key in sorted(cluster_mapping.keys())])

plt.title('DBSCAN Clustering after PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Step 1: Apply PCA to reduce to 2 components
pca = PCA(n_components=2)
pca_data = pca.fit_transform(normalized_data)

# Step 2: Update cluster mapping dynamically based on DBSCAN labels
unique_labels = sorted(set(dbscan_labels))  # Find unique DBSCAN labels
cluster_mapping = {
    -1: "Outliers",  # Default label for outliers
    0: "Cluster 0 - Developed Countries",
    1: "Cluster 1 - Developing Countries",
    2: "Cluster 2 - Least Developed Countries",
    # Add labels for any unexpected clusters
    **{label: f"Cluster {label}" for label in unique_labels if label not in [-1, 0, 1, 2]},
}

# Map DBSCAN labels to descriptive names
descriptive_labels = [cluster_mapping[label] for label in dbscan_labels]

# Step 3: Visualize PCA-reduced data with descriptive cluster labels
plt.figure(figsize=(10, 6))
scatter = plt.scatter(
    pca_data[:, 0],
    pca_data[:, 1],
    c=dbscan_labels,
    cmap='viridis',
    s=100,
)

# Add color bar with cluster descriptions
cbar = plt.colorbar(scatter, ticks=unique_labels)
cbar.ax.set_yticklabels([cluster_mapping[label] for label in unique_labels])

plt.title('DBSCAN Clustering after PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Apply PCA to reduce to 2 components
pca = PCA(n_components=2)
pca_data = pca.fit_transform(normalized_data)

# Step 2: Define descriptive labels for the selected clusters (3 clusters)
selected_clusters = [0, 1, 2]  # Select the 3 clusters you want to visualize
cluster_mapping = {
    -1: "Outliers",
    0: "Cluster 0 - Developed Countries",
    1: "Cluster 1 - Developing Countries",
    2: "Cluster 2 - Least Developed Countries"
}

# Filter DBSCAN labels to include only the selected clusters
filtered_indices = np.isin(dbscan_labels, selected_clusters)
filtered_pca_data = pca_data[filtered_indices]
filtered_labels = np.array(dbscan_labels)[filtered_indices]

# Step 3: Update mapping for the selected clusters
filtered_cluster_mapping = {key: cluster_mapping[key] for key in selected_clusters}

# Step 4: Visualize PCA-reduced data with descriptive cluster labels
plt.figure(figsize=(10, 6))
scatter = plt.scatter(
    filtered_pca_data[:, 0], 
    filtered_pca_data[:, 1], 
    c=filtered_labels, 
    cmap='viridis', 
    s=100
)

# Add color bar with cluster descriptions
cbar = plt.colorbar(scatter, ticks=sorted(filtered_cluster_mapping.keys()))
cbar.ax.set_yticklabels([filtered_cluster_mapping[key] for key in sorted(filtered_cluster_mapping.keys())])

plt.title('DBSCAN Clustering after PCA (3 Clusters)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
